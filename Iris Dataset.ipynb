## Task 1 - Data Preprocessing

### Import

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

### Load the dataset

```python
df = pd.read_csv('iris.csv')
df.head()
```

### Check for missing values

```python
df.isnull().sum()
```

### Data visualization - pairplot

```python
sns.pairplot(df, hue='species', height=2, aspect=1)
plt.show()
```

### Label Encoding for categorical target

```python
le = LabelEncoder()
df['species'] = le.fit_transform(df['species'])  # setosa=0, versicolor=1, virginica=2
```

### Feature-target split

```python
X = df.drop('species', axis=1)
y = df['species']
```

```python
X
```

```python
y
```

### Standardization

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

### Split into train/test

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

```python
X_test
```

```python
y_train
```

```python
y_test
```

## Task 3 - KNN Classifier

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

### Try different K values

```python
for k in [3, 5, 7]:
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)
    y_pred_knn = knn.predict(X_test)
    
    print(f"\n------------------------------ K = {k} ------------------------------")
    print("Accuracy:", accuracy_score(y_test, y_pred_knn))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_knn))
    print("Classification Report:\n", classification_report(y_test, y_pred_knn))
```

# Level - 2

## Task 1 - Logistic Regression

### Filter only 2 classes

```python
df_binary = df[df['species'].isin([0, 1])]
X_bin = df_binary.drop('species', axis=1)
y_bin = df_binary['species']
```

### Train-test split

```python
X_train_bin, X_test_bin, y_train_bin, y_test_bin = train_test_split(X_bin, y_bin, test_size=0.2, random_state=42)
```

### Model

```python
from sklearn.linear_model import LogisticRegression
log_reg = LogisticRegression()
log_reg.fit(X_train_bin, y_train_bin)
```

### Predictions and evaluation

```python
from sklearn.metrics import roc_curve, auc
y_pred_bin = log_reg.predict(X_test_bin)
y_prob = log_reg.predict_proba(X_test_bin)[:, 1]
```

```python
accuracy_score(y_test_bin, y_pred_bin)
```

```python
print(classification_report(y_test_bin, y_pred_bin))
```

### ROC Curve

```python
fpr, tpr, _ = roc_curve(y_test_bin, y_prob)
roc_auc = auc(fpr, tpr)

plt.plot(fpr, tpr, label=f'AUC = {roc_auc:.2f}')
plt.plot([0, 1], [0, 1], linestyle='--')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Logistic Regression")
plt.legend()
plt.grid(True)
plt.show()
```

## Task 3 - K-Means Clusturing

```python
from sklearn.cluster import KMeans
```

### Elbow Method

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)
```

### Plot Elbow

```python
plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Method for Optimal k")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.grid(True)
plt.show()
```

### Apply KMeans

```python
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(X_scaled)
```

### Plot Cluster

```python
plt.scatter(X_scaled[:, 0], X_scaled[:, 1], c=clusters, cmap='viridis')
plt.title("K-Means Clustering")
plt.xlabel("Sepal Length (scaled)")
plt.ylabel("Sepal Width (scaled)")
plt.grid(True)
plt.show()
```

## Task 2 - SVM

```python
from sklearn.svm import SVC
```

### Try linear and RBF kernels

```python
for kernel in ['linear', 'rbf']:
    svm = SVC(kernel=kernel, probability=True)
    svm.fit(X_train, y_train)
    y_pred_svm = svm.predict(X_test)
    
    print(f"\n------------------------------ SVM with {kernel} kernel ------------------------------")
    print("Accuracy:", accuracy_score(y_test, y_pred_svm))
    print("Classification Report:\n", classification_report(y_test, y_pred_svm))
```

### Visualize decision boundaries

```python
from mlxtend.plotting import plot_decision_regions
```

```python
X_vis = X_scaled[:, :2]
svm_vis = SVC(kernel='rbf')
svm_vis.fit(X_vis, y)
```

```python
plot_decision_regions(X_vis, y.values, clf=svm_vis, legend=2)
plt.title("SVM Decision Boundary (RBF Kernel)")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

## Task 3 - Neural Network with TensorFlow/Keras

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Input
```

### Model architecture

```python
model = Sequential([
    Input(shape=(4,)),
    Dense(10, activation='relu'),
    Dense(8, activation='relu'),
    Dense(3, activation='softmax')
])
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

### Train the model

```python
history = model.fit(X_train, y_train, epochs=50, validation_data=(X_test, y_test), verbose=0)
```

### Evaluate

```python
loss, acc = model.evaluate(X_test, y_test)
```

### Plot training history

```python
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title("Neural Network Accuracy")
plt.xlabel("Epoch")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```
